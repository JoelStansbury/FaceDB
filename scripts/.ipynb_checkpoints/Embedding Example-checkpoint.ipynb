```python
import cv2 #https://pypi.org/project/opencv-python/
import os
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import faceFinder
```

```python
result = faceFinder.process_images('F:\Pictures')
```

```python
import numpy as np
X = np.array(result['face_vectors'])
X.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 200)
person = kmeans.fit_predict(X)
```

```python
plt.hist(person,bins=200)
plt.show()
```

```python
imgs = result['fnames']
bboxs = result['bounding_boxes']

for i,c in enumerate(person):
    if c==5:
        img = cv2.cvtColor(cv2.imread(imgs[i]), cv2.COLOR_BGR2RGB)
        plt.imshow(img)
        ax = plt.gca()
        x,y,w,h = bboxs[i]
        rect = Rectangle((x,y),w,h,linewidth=1,edgecolor='r',facecolor='none')
        ax.add_patch(rect)
        plt.show()
```
